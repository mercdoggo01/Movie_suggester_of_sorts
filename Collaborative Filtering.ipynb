### Step 5: Recommending Similar Movies

#### We will be exploring two methods here but fundamentals of both are same both rely on a matrix of userid vs movie title and will be selected on basis of number of ratings

#### Creating a matrix on the basis of what rating was given to that particular movie

```python
mat = movies_final_df.pivot_table(index = 'userId',columns='title',values='rating')
mat.head()
```

#### Most rated movies

```python
highest_ratings_df.sort_values('number_of_ratings', ascending=False).head(10)
```

## Method 1:

```python
Forrest_Gump_user_ratings = mat['Forrest Gump (1994)']
Forrest_Gump_user_ratings
```

```python
similar_to_Forrest  = mat.corrwith(Forrest_Gump_user_ratings)
```

#### Removal of NaN values and using Dataframe instead of series

```python
corr_Forr = pd.DataFrame(similar_to_Forrest,columns=['correlation'])
corr_Forr.dropna(inplace=True)
corr_Forr.head()
```

```python
corr_Forr.sort_values('correlation',ascending=False).head(10)
```

#### The above results doesn't make sense as alot of movies were watched by users who watched Forrest Gump(perks of being a popular movie)

#### <span style="color:red">Solution to this issue:</span> Filter movies based on number of ratings 

```python
corr_Forr = corr_Forr.join(highest_ratings_df['number_of_ratings'])
corr_Forr.head()
```

```python
corr_Forr[corr_Forr['number_of_ratings']>50].sort_values('correlation',ascending=False).head(20)
```

### Method 2:

#### We will drop all those movies with very few user ratings

#### Reason behind this being the user won't be suggested less watched movies

```python
user_ratings = mat.dropna(thresh=10,axis=1).fillna(0)
# filling non-rated with 0's
```

```python
user_ratings
```

#### Using inbuilt pearson correlation

```python
# Similarity matrix
item_similarity_df = user_ratings.corr(method='pearson')
item_similarity_df
```

```python
# Recommendation process

def get_similar(movie_name,user_rating):
    similar_score = item_similarity_df[movie_name]*(user_rating-2.5)
    similar_score = similar_score.sort_values(ascending=False)
    
    return similar_score

#print(get_similar("xyz",1))
```

```python
movie_list = [("Forrest Gump (1994)",5),("Schindler's List (1993)",3),("Terminator 2: Judgment Day (1991)",2)]

similar_movies = pd.DataFrame()

for movie,rating in movie_list:
    similar_movies = similar_movies.append(get_similar(movie,rating),ignore_index=True)
    
similar_movies.head()
#similar_movies.sum().sort_values(ascending=False)
```
